## `array()` method: signal over many intervals in parallel

* Example: *"what's the signal under each peak?"*
* Get a NumPy array of signal, where each row is the signal for one peak
* Peaks are not the same size, so use `bins=100` to scale each to 0-100% over 100 bins
* Use 8 cores, `multiprocessing.Pool.map`
    * See `metaseq/scripts/speedtest.py` for evaluating how well it scales with your hardware


```python
bam = metaseq.genomic_signal('data/G1E-ER_Gata1_chr7.bam', 'bam')
import pybedtools
windows = pybedtools.BedTool('data/G1E-ER_Gata1_chr7.broadPeak')
a = bam.array(windows, bins=100, processes=8)
print "array shape:", a.shape
```

## Plot array

Choosing colormaps and row order can be tricky; see docs for some options


```python
# Normalize
s = a.std(axis=1)[:, newaxis]; row_mean = a.mean(axis=1)[:, newaxis]
b = (a - row_mean) / s

# Sort by position of max signal within a peak ("summit")
maxpos = argmax(b, axis=1);
ind = argsort(maxpos)

imshow(b[ind, :], origin='lower')
cbar = colorbar(); cbar.set_label('Normalized read density per bin')
axvline(25, color='k', linestyle=':'); axvline(75, color='k', linestyle=':');
xlabel('bins within peak'); ylabel('peaks');
```

## Same data, different row order

```python
# Sort by TIP score (Cheng et al. 2011, Bioinformatics 27(23):3221-3227)
ind = argsort(metaseq.plotutils.tip_zscores(b))

imshow(b[ind, :], origin='lower', interpolation='nearest', aspect='auto', cmap=cm.Reds)
cbar = colorbar(); cbar.set_label('Normalized read density per bin'); 
xlabel('bins within peak'); ylabel('peaks');
```

## Plot summary ("meta-peak")

```python
def ci(a):
    "Column-wise 95% CI"
    return 1.96 * a.std(axis=0) / np.sqrt(len(a))

fig=figure(); ax=fig.add_subplot(111);
x = range(0, 100)
col_mean = a.mean(axis=0)
ax.set_ylabel('avg read density per bin')
fill_between(x, col_mean-ci(a), col_mean+ci(a), alpha=0.1, edgecolor='None')
plot(a.mean(axis=0), '0.5'); xlabel('bins');
```

# Clustering peaks

Using the same API, but different data: *"how do peaks cluster around TSSes?"*

```python
# Previous example:
# bam = metaseq.genomic_signal('data/G1E-ER_Gata1_chr7.bam', 'bam')
# windows = pybedtools.BedTool('data/G1E-ER_Gata1_chr7.broadPeak')
# arr = bam.array(windows, bins=100, processes=8)

bed = metaseq.genomic_signal('data/G1E-ER_Gata1_chr7.bigbed', 'bigbed')
windows = pybedtools.BedTool('data/TSS.gff').slop(genome='mm9', b=5000)
arr = bed.array(windows, bins=100, processes=8)

# Discard genes with no peaks
has_peak = arr.sum(axis=1)>0
arr = arr[has_peak, :]
```

## Plot clustered array

```python
# Cluster
ind, breaks = metaseq.plotutils.clustered_sortind(
    arr, k=6, scorefunc=lambda x: argmax(x.sum(axis=0)))

figure(figsize=(3, 5));
imshow(arr[ind, :], aspect='auto', interpolation='nearest', cmap=matplotlib.cm.binary, origin='lower', extent=(-5, 5, 0, len(arr)));
for b in breaks:
    axhline(b, color='r', linestyle=':')
axvline(0, color='k', linestyle=':')
xlabel('distance from TSS (kbp)'); ylabel('TSS with peak nearby');
```

# `metaseq.results_table`

* Loose wrapper around `pandas.DataFrame`; accessible as the `.data` attribute
* `metaseq.results_table.ResultsTable` is generic (any table keyed by ID)
* `metaseq.results_table.DESeqResults` is specific to output of a commonly-used `R` package

```python
from metaseq.results_table import DESeqResults
d = DESeqResults('data/deseq_results.txt',
    # Contains ID -> interval mapping:
    db='data/Mus_musculus.NCBIM37.67_chr7_cleaned.gtf.db')
d = d.strip_unknown_features(); d
```

# Mini-browsers as callbacks

"mini-browsers" have a plugin-like architecture, and are designed to operate as callbacks.

Set one up, and call the `plot()` method with an interval

```python
from metaseq.minibrowser import SignalMiniBrowser
mb = SignalMiniBrowser(genomic_signal_objs=[bam])
def callback(i):
    mb.plot(i); fix(gca()); gca().set_xlabel('chr7')

# USAGE:
# d.scatter(..., callback=callback, ...)

# *CLICK*
callback(d.db['ENSMUSG00000030403'])
```

## A `GeneModelMiniBrowser`

```python
from metaseq.minibrowser import GeneModelMiniBrowser
control = metaseq.genomic_signal('data/G1E-ER_Input_chr7.bam', 'bam')
mb = GeneModelMiniBrowser(
        [bam, control], db=d.db,
        plotting_kwargs=[dict(color='r'), dict(color='.5')])

def callback(i):
    interval = d.db[i]
    interval.start -= 1000; interval.stop += 1000
    mb.plot(interval)
    xlabel(interval.chrom); axis(xmin=interval.start, xmax=interval.stop)

# *CLICK*
callback('ENSMUSG00000030403')
```

## Average reads per bin

```python
fig = figure(figsize=(10,6)); 
for i, nbin in enumerate([100, 500, 1000, None]):
    ax = fig.add_subplot(4, 1, i + 1)
    #-------------------------------------------
    x, y = bam.local_coverage(window, bins=nbin) 
    #-------------------------------------------
    ax.plot(x, y, '0.5'); ax.text(.9, .8, "bins=%s" % nbin, transform=ax.transAxes)
    ax.set_ylabel('avg. reads\nper bin')
for ax in fig.axes[:-1]:
    ax.xaxis.set_visible(False); 
```

## Total reads per bin
(note y-axis, `preserve_total=True` kwarg)

```python
fig = figure(figsize=(10,6)); 
for i, nbin in enumerate([100, 500, 1000, None]):
    ax = fig.add_subplot(4, 1, i + 1)
    #----------------------------------------------------------------
    x, y = bam.local_coverage(window, bins=nbin, preserve_total=True)     
    #----------------------------------------------------------------
    ax.plot(x, y, '0.5'); ax.set_ylabel('total reads\nper bin')
    ax.text(.9, .8, "bins=%s" % nbin, transform=ax.transAxes)
for ax in fig.axes[:-1]:
    ax.xaxis.set_visible(False);
```
